```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html,"html.parser")
```

```python
print(soup.text)
```

```python
title = []
body = []
tags = []


for i in soup.find_all("div",class_="s-post-summary--content"):
    
    # title
    text = i.find("h3",class_="s-post-summary--content-title").text
    regex = re.findall(r"\n(.*?)\n",text)
    if regex:
        title.append(regex[0])
    else:
        title.append(np.nan)
        
    # body
    text = i.find("div",class_="s-post-summary--content-excerpt").text.strip()
    regex = re.sub(r"http\S+",'',text)
    regex = re.sub(r"\s+",' ',regex).strip()
    if regex:
        body.append(regex)
    else:
        body.append(np.nan)
        
    # tags
    text = i.find("ul",class_="ml0 list-ls-none js-post-tag-list-wrapper d-inline")
    if text:
        tags.append(text.text.strip())
    else:
        tags.append(["No tags found!"])

```

```python
print("Title: ",len(title))
print("Body: ",len(body))
print("Tags: ",len(tags))
```

```python
df = pd.DataFrame({"Title":title,"Body":body,"Tags":tags})
```

```python
def pre_proce(data,coln,case="lower",tages=True,urls=True,email=True,mentions=True,digits=True,dates=True,emojii=True,contractionss=True,stopwordss=True,infl="stem",stemms="poter",punc=True):
    import re
    import emoji
    import contractions
    import nltk
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
    ps = PorterStemmer()
    ls = LancasterStemmer()
    ss = SnowballStemmer(language = "english")
    wl = WordNetLemmatizer()

    # stopwords creation
    stp = stopwords.words("english")
    stp.remove("not")

    # emoji
    if emojii == True:
        data[coln]=data[coln].apply(lambda x: emoji.demojize(x,delimiters=('','')))
    else:
        pass

    # lower and upper case
    if case == "lower":
        data[coln] = data[coln].str.lower()
    elif case == "upper":
        data[coln] = data[coln].str.upper()
    else:
        pass

    # tages
    if tages == True:
        data[coln]=data[coln].apply(lambda x: re.sub("<.*?>"," ",x))
    else:
        pass

    # urls
    if urls == True:
        data[coln]=data[coln].apply(lambda x: re.sub("https?://\S+"," ",x))
    else:
        pass

    # emails
    if email == True:
        data[coln] = data[coln].apply(lambda x: re.sub("\S+@\S+"," ",x))
    else:
        pass

    # mentions
    if mentions == True:
        data[coln] = data[coln].apply(lambda x:re.sub("\B[@#]\S+"," ",x))
    else:
        pass

    # digits
    if digits == True:
        data[coln]=data[coln].apply(lambda x: re.sub("\d"," ",x))
    else:
        pass

    # dates
    if dates == True:
        data[coln]=data[coln].apply(lambda x: re.sub(r"\b[0-9]{4}\/[0-9]{1,2}\/[0-9]{1,2}"," ",x))
        data[coln]=data[coln].apply(lambda x: re.sub(r"\b[0-9]{1,2}\/[0-9]{1,2}\/[0-9]{4}$"," ",x))
    else:
        pass

    # contraction
    if contractionss == True:
        data[coln]=data[coln].apply(lambda x: contractions.fix(x))
    else:
        pass

  

    # punctuations
    if punc == True:
        data[coln] = data[coln].apply(lambda x: re.sub('[!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]'," ",x))
    else:
        pass
                    

    return data
```
